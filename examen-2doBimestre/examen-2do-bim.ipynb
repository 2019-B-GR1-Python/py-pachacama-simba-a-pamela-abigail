7. Crear un Dataframe con 10 registros y 6 columnas y Ordenar el dataframe 1 vez por cada columna, ascendente y descendente

```python
arr_pand6 =np.random.randint(0,100,60).reshape(10,6)
df6 = pd.DataFrame(arr_pand6)
df6
```

```python
for i in range(0,6):
    print(df6.sort_values(i))
    print("------------------------------")
```

```python
for i in range(0,6):
    print(df6.sort_values(i, ascending=False))
    print("------------------------------")
```
