```python
from os import remove
class Gestionar:
    ingrediente =" "
    def __init__(self):
        print("--------------------------------------------------")
        print("Comida")
        print("--------------------------------------------------")
        self.menu_padre()
    

        
    def abrir_padres(self):
        try:
            path = "./padres.txt"
            archivo_abierto= open(path)
            contenido = archivo_abierto.readlines()
            for linea in contenido:
                print(linea)
                
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir padre")
            
    def new_padre(self):
            try:
                path = "./padres.txt"
                archivo_abierto= open(path, mode="a")
                archivo_abierto.write("\nNombre:"+input("Nombre:")+",Precio:"+input("Precio: ")+", Calificacion:"
                                      +input("Calificacion: "))
                archivo_abierto.close()
            except Exception as error:
                print("Error nuevo padre")
                
    def actualizar_padres(self):
        try:
            path = "./padres.txt"
            archivo_abierto= open(path,"r")
            lineas = archivo_abierto.readlines()
            numero=int(input("Padre:"))
            contenido = lineas[numero - 1]
            print(contenido)
            archivo_abierto= open(path,"w")
            print("Actualizar padre")
            
            for linea in lineas:
                if linea!=contenido:
                    archivo_abierto.write(linea)
                elif linea ==  contenido:
                    archivo_abierto.write("Nombre:"+input("Nombre:")+",Precio:"+input("Precio: ")+", Calificacion:"
                                  +input("Calificacion: ")+"\n")            
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir padre")
    
    def eliminar_padres(self):
        try:
            path = "./padres.txt"
            archivo_abierto= open(path,"r")
            lineas = archivo_abierto.readlines()
            numero=int(input("Padre:"))
            contenido = lineas[numero - 1]
            print(contenido)
            nombre=input("Confirme el nombre:")
            archivo_abierto= open(path,"w")
            print("Padre Eliminado")
            
            for linea in lineas:
                if linea!=contenido:
                    archivo_abierto.write(linea)
                    ##remove(nombre+"-hijos.txt")
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir padre")
    
    def abrir_hijo(self):
        try:
            print(self.ingrediente)
            path = "./"+self.ingrediente+"-hijos.txt"
            print(path)
            archivo_abierto= open(path,"r+")
            contenido = archivo_abierto.readlines()
            for linea in contenido:
                print(linea)
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir hijo")
            
    def new_hijo(self):
            try:
                path = "./"+self.ingrediente+"-hijos.txt"
                archivo_abierto= open(path, mode="a")
                archivo_abierto.write("\nNombre:"+input("Nombre:")+",Cantidad:"+input("Peso: ")+", Precio:"+input("Precio: "))
                archivo_abierto.close()
            except Exception as error:
                print("Error nuevo hijo")
    
    def actualizar_hijos(self):
        try:
            path = "./"+self.ingrediente+"-hijos.txt"
            archivo_abierto= open(path,"r")
            lineas = archivo_abierto.readlines()
            numero=int(input("Hijo:"))
            contenido = lineas[numero - 1]
            print(contenido)
            archivo_abierto= open(path,"w")
            print("Actualizar hijo")
            
            for linea in lineas:
                if linea!=contenido:
                    archivo_abierto.write(linea)
                elif linea ==  contenido:
                    archivo_abierto.write("Nombre:"+input("Nombre:")+",Cantidad:"+input("Peso: ")+", Precio:"
                                          +input("Precio: ")+"\n")            
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir hijo")
    
    def eliminar_hijos(self):
        try:
            path = "./"+self.ingrediente+"-hijos.txt"
            archivo_abierto= open(path,"r")
            lineas = archivo_abierto.readlines()
            numero=int(input("Ingrediente #:"))
            contenido = lineas[numero - 1]
            print(contenido)
            archivo_abierto= open(path,"w")
            print("Hijo Eliminado")
            
            for linea in lineas:
                if linea!=contenido:
                    archivo_abierto.write(linea)
                
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir hijo")
    
    def menu_padre(self):
        print("--------------------------------------------------")
        print("Operaciones")
        print("--------------------------------------------------")
        print("1. nuevo Padre")
        print("2. ver Padre")
        print("3. actualizar Padre")
        print("4. eliminar Padre")
        print("5. gestionar hijo")
        opcion=int(input("Escoja una operacion:"))
        if(opcion == 1):
            self.new_padre()
            print("Comida registrado con exito")
            self.menu_padre()
        elif(opcion == 2):
            self.abrir_padres()
            self.menu_padre()
        elif(opcion == 3):
            self.actualizar_padres()
            self.menu_padre()
        elif(opcion == 4):
            self.eliminar_padres()
            self.menu_padre()
        elif(opcion == 5):
            self.ingrediente = input("Nombre Padre:")
            print(self.ingrediente)
            self.menu_hijo()
        
    def menu_hijo(self):
        print("--------------------------------------------------")
        print("Operaciones")
        print("--------------------------------------------------")
        print("1. nuevo hijo")
        print("2. ver hijos")
        print("3. actualizar hijos")
        print("4. eliminar hijos")
        print("5. menu padre")
        opcion=int(input("Escoja una operacion:"))
        if(opcion == 1):
            self.new_hijo()
            print("Ingrediente registrado con exito")
            self.menu_hijo()
        elif(opcion == 2):
            self.abrir_hijo()
            self.menu_hijo()
        elif(opcion == 3):
            self.actualizar_hijos()
            self.menu_hijo()
        elif(opcion == 4):
            self.eliminar_hijos()
            self.menu_hijo()
        elif(opcion == 5):
            self.menu_padre()
        
```
