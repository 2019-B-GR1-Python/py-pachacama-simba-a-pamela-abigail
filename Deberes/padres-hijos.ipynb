```python
class Gestionar:
    def __init__(self):
        print("--------------------------------------------------")
        print("Comida")
        print("--------------------------------------------------")
        self.menu_padre()
    

        
    def abrir_padres(self):
        try:
            path = "./padres.txt"
            archivo_abierto= open(path)
            contenido = archivo_abierto.readlines()
            for linea in contenido:
                print(linea)
                
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir padre")
            
    def new_padre(self):
            try:
                path = "./padres.txt"
                archivo_abierto= open(path, mode="a")
                archivo_abierto.write("\nNombre:"+input("Nombre:")+",Precio:"+input("Precio: "))
                archivo_abierto.close()
            except Exception as error:
                print("Error nuevo padre")
    
    def abrir_hijo(self):
        try:
            path = "./tor-hijos.txt"
            archivo_abierto= open(path)
            contenido = archivo_abierto.readlines()
            for linea in contenido:
                print(linea)
            archivo_abierto.close()
        except Exception as error:
            print("Error abrir padre")
            
    def new_hijo(self):
            try:
                path = "./tor-hijos.txt"
                archivo_abierto= open(path, mode="a")
                archivo_abierto.write("\nNombre:"+input("Nombre:")+",Cantidad:"+input("cantidad: "))
                archivo_abierto.close()
            except Exception as error:
                print("Error nuevo padre")
    
    def menu_padre(self):
        print("--------------------------------------------------")
        print("Operaciones")
        print("--------------------------------------------------")
        print("1. nuevo Padre")
        print("2. ver Padre")
        print("3. gestionar hijo")
        opcion=int(input("Escoja una operacion:"))
        if(opcion == 1):
            self.new_padre()
            print("Comida registrado con exito")
            self.menu_padre()
        elif(opcion == 2):
            self.abrir_padres()
            self.menu_padre()
        elif(opcion == 3):
            self.menu_hijo()
        
    def menu_hijo(self):
        print("--------------------------------------------------")
        print("Operaciones")
        print("--------------------------------------------------")
        print("1. nuevo hijo")
        print("2. ver hijos")
        print("3. menu padre")
        opcion=int(input("Escoja una operacion:"))
        if(opcion == 1):
            self.new_hijo()
            print("Ingrediente registrado con exito")
            self.menu_hijo()
        elif(opcion == 2):
            self.abrir_hijo()
            self.menu_hijo()
        elif(opcion == 3):
            self.menu_padre()
        
```

```python
nuevo_padre = Gestionar()
print(nuevo_padre)
```
