* Analizar los cantones con mayor numero de fallecidos.

```python
data_accident_trns=data.loc[:,'NUM_FALLECIDO'].groupby(data.loc[:,'CANTON']).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_accident_trns.head(6).index,data_accident_trns.head(6),palette='coolwarm')
```

**Resultados:** El canton que tiene mayor numero de fallecidos es Pelenque perteneciente a la provincia de Los Rios.

* Analizar el número de fallecidos de acuerdo a la zona.

```python
serie_fallecidos=data.ZONA.value_counts()
total_accidentes=data.shape[0]
porcentaje_rural=(serie_fallecidos[2]/total_accidentes)*100
porcentaje_urbano=(serie_fallecidos[1]/total_accidentes)*100
patches,texts=plt.pie(serie_fallecidos,shadow=True,startangle=90)
plt.pie(serie_fallecidos,shadow=True,labels=serie_fallecidos.index,autopct='%1.1f%%')
plt.legend(patches,serie_fallecidos.index,loc="best")
plt.axis("equal")
plt.title("Fallecidos de acuerdo a la zona")
plt.tight_layout()
numero_fallecidos_rural=serie_fallecidos[2]
numero_fallecidos_urbano=serie_fallecidos[1]
plt.show()
```

**Resultados:** En el año 2016, la mayor parte de fallecidos se danen la zona urbana con un 71,5%

* Determinar el número de lesionados por día

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(data.groupby(['DIA']).NUM_LESIONADO.sum())
ax.set_title('LESIONADOS POR DIA DE LA SEMANA')
ax.set_xlabel('DIA')
ax.set_ylabel('NUMERO_LESIONADOS')
```

**Resultado:** La mayor parte de lesionados se dan los días sábados y domingos

* Analizar la hora con mayor victimas.

```python
serie_horas=data.HORA.value_counts().head(6)
plt.pie(serie_horas,shadow=True,labels=serie_horas.index,autopct='%1.1f%%')
plt.legend(patches,serie_horas.index,loc="best")
plt.axis('equal')
plt.title("Horas de accidentes")
plt.tight_layout()
plt.show()
```

**Resultado:** La hora en donde existen mas victimas es a las 19 con un 17,6%, seguido por las 17 horas con un 16,7%
